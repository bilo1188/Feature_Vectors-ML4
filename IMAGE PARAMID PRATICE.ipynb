```python
#import important libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python

%matplotlib inline

#taking image as input from directory
img = cv2.imread('images/rainbow_flag.jpg')

#change color from bge to RGB
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#for ploting img
plt.imshow(img)
```

```python
level1 = cv2.pyrDown(img)
level2 = cv2.pyrDown(level1)
level3 = cv2.pyrDown(level2)

#display the img

f, (ax1,ax2,ax3,ax4) = plt.subplots(1,4 , figsize=(20,30))

ax1.imshow(img)
ax1.set_title('orginal image')

ax2 . imshow(level1)
ax2.set_xlim([0, img.shape[1]])
ax2.set_ylim([0, img.shape[0]])

ax3 . imshow(level2)
ax3.set_xlim([0, img.shape[1]])
ax3.set_ylim([0, img.shape[0]])

ax4 . imshow(level3)
ax4.set_xlim([0, img.shape[1]])
ax4.set_ylim([0, img.shape[0]])
```
